```python
import pandas as pd
import sqlite3
```

## create a connection to the database

```python
conn= sqlite3.connect('../data/checking-logs.sqlite')
```

## get the schema of the table checker

```python
pd.io.sql.read_sql('PRAGMA table_info(checker)', conn)
```

## get only the first 10 rows of the table checker

```python
pd.io.sql.read_sql('select * from checker limit 10', conn)
```

## count how many rows satisfy the following conditions: count the rows from the pageviews table but only with users from the checker table with:
status = ’ready’;
numTrials = 1; 
labnames should be from the list: ’laba04’, ’laba04s’, ’laba05’, ’laba06’, ’laba06s’, ’project1’; 
store in the dataframe checkers with the column cnt;


```python
query = """
SELECT COUNT(*) AS cnt
FROM pageviews
WHERE uid IN (
    SELECT uid
    FROM checker
    WHERE status = 'ready'
      AND numTrials = 1
      AND labname IN ('laba04', 'laba04s', 'laba05', 'laba06', 'laba06s', 'project1')
);
"""
checkers = pd.read_sql(query, conn)
checkers
```
