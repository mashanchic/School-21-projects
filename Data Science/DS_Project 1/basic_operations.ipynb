## create a dataframe views with two columns: datetime and user

```python
import pandas as pd
views = pd.read_csv('../data/feed-views.log',
                 names=['datetime', 'user'],
                 sep='\t')
views['datetime'] = pd.to_datetime(views['datetime'])
views.info()
views['year'] = views['datetime'].apply(lambda x: x.year)
views['month'] = views['datetime'].apply(lambda x: x.month)
views['day'] = views['datetime'].apply(lambda x: x.day)
views['hour'] = views['datetime'].apply(lambda x: x.hour)
views['minute'] = views['datetime'].apply(lambda x: x.minute)
views['second'] = views['datetime'].apply(lambda x: x.second)
views
```

## create the new column daytime 

## 0 – 3.59 night, 4 – 6.59 early morning, 7 – 10.59 morning, 11 – 16.59 afternoon, 17 – 19.59 early evening, 20 – 23.59 evening

```python
bins = [-1, 3.99, 6.99, 10.99, 16.99, 19.99, 23.99] 
labels = ['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
views['daytime'] = pd.cut(views['hour'], bins=bins, labels=labels)
views.set_index('user', inplace=True)
views
```

```python
views.info()
```

## calculate the number of elements in your dataframe

```python
total_elements = views.count().sum()
print(f"Total number of elements: {total_elements}")
daytime_counts = views['daytime'].value_counts()
print("\nNumber of elements in each time of day category:")
daytime_counts
```

## sort values in the dataframe

```python
views_sorted = views.sort_values(by=['hour', 'minute', 'second'])
views_sorted
```

## calculate min and max for hours and mode for the daytime categories

```python
max_hour_night = views[views['daytime'] == 'night']['hour'].max()
min_hour_morning = views[views['daytime'] == 'morning']['hour'].min()
mode_hour = views['hour'].mode()[0]
mode_daytime = views['daytime'].mode()[0]
print(f"Maximum for night hours {max_hour_night};",f"Minimum for morning hours {min_hour_morning};",f"Mode for the daytime {mode_daytime};",f"Mode for the hour {mode_hour}")
```

## 3 earliest/latest hours in the morning

```python
morning_views = views[views['daytime'] == 'morning']
earliest_morning_hours = morning_views.nsmallest(3, 'hour')
earliest_morning_hours
```

```python
latest_morning_hours = morning_views.nlargest(3, 'hour')
latest_morning_hours
```

```python
stats = views['hour'].describe()
stats
```
