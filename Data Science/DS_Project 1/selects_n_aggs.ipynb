```python
import pandas as pd
```

## load the JSON file

```python
df = pd.read_json('../data/auto.json',
                 orient='records')
df.set_index('CarNumber', inplace=True)
df
```

## display the rows only where the fines are more than 2,100

```python
fines_over_2100=df[df['Fines'] > 2100]
fines_over_2100
```

## display the rows only where the fines are more than 2,100 and the refund equals 2

```python
fines_over_2100_refund_2 = df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
fines_over_2100_refund_2
```

## display the rows only where the models are from the list: [’Focus’, ’Corolla’]

```python
models_list = ['Focus', 'Corolla']
models_selected = df[df['Model'].str.contains('|'.join(models_list), na=False)]
models_selected
```

## display the rows only where the car number is from the list: [’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’]

```python
car_numbers= ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
car_numbers_selected = df[df.index.isin(car_numbers)]
car_numbers_selected
```

## make the aggregations with the make and the model 

## display the median fines grouped by the make

```python
median_fines_make = df.groupby('Make')['Fines'].median()
median_fines_make
```

## display the median fines grouped by the make and the model

```python
median_fines_make_model = df.groupby(['Make', 'Model'])['Fines'].median()
median_fines_make_model
```

## display the number of fines grouped by the make and the model

```python
count_fines = df.groupby(['Make', 'Model'])['Fines'].count()
count_fines
```

## display the minimum and the maximum fines grouped by the make and the model

```python
min_max_fines = df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
min_max_fines
```

## display the standard deviation of the fines grouped by the make and the model

```python
std_fines = df.groupby(['Make', 'Model'])['Fines'].std()
std_fines
```

## make the aggregations with the car number

## display the car numbers grouped by the number of the fines in descending order

```python
car_numbers_fines = df.groupby(df.index)['Fines'].count().sort_values(ascending=False)
car_numbers_fines
```

```python
car_fines_sum = df.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False)
car_fines_sum
```
