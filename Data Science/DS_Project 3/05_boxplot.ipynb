```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

## make a connection to the databaseÂ¶

```python
df = pd.read_csv('../data/ab-test.csv')
df
```

## devide for two dataframes

```python
df_before = df[df['time'] == 'before']
df_after = df[df['time'] == 'after']
df_before.info()
df_after.info()
```

## create two boxplots

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8), sharey = True)
df_after.boxplot(column='diff', 
                                  by='group', 
                                  ax=ax[0], 
                                  color=dict(whiskers='red',
                                             medians='red', caps='red'), 
                                  boxprops=dict(color = 'green', linewidth=3),
                                  medianprops=dict(linewidth=2)) 
df_before.boxplot(column='diff',
                                   by='group',
                                   ax=ax[1],
                                   color=dict(whiskers='red',
                                              medians='red', caps='red'),
                                   boxprops=dict(color = 'green', linewidth=3),
                                   medianprops=dict(linewidth=2))
ax[0].set_title('after')
ax[1].set_title('before')
fig.suptitle(
    'Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
plt.show()
```
